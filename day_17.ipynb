```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

nltk.download(['stopwords','wordnet','punkt'],quiet=True)
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess(text):
   text=text.lower()

   text=re.sub(r'[^a-z]',' ',text)
   tokens=text.split()
   filtered_tokens =[]
   for w in tokens:
    if w not in stop_words:
      filtered_tokens.append(w)
   tokens=filtered_tokens

   lemmatized_tokens=[]
   for w in tokens:
    lemmatized_tokens.append(lemmatizer.lemmatize(w))
   tokens = lemmatized_tokens
   return ' '.join(tokens)
print(preprocess('I am walking towards the  nearest park to see the Visuals'))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
email={
    'cricket match cricket match ',
    'free win money free  win money',
    'cricket match free win money cricket'
}
cv=CountVectorizer()
x=cv.fit_transform(email).toarray()
print('vocabullary: ',cv.get_feature_names_out())
print(x)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
email={
    'cricket match cricket match ',
    'free win money free  win money',
    'cricket match free win money cricket'
}
tfidf=TfidfVectorizer( sublinear_tf=True,max_df=0.95,min_df=1)
x=tfidf.fit_transform(email).toarray()
print('feature names: ',tfidf.get_feature_names_out())
print('tfidf matrix')
import numpy as np
print(np.round(x,3))
```
