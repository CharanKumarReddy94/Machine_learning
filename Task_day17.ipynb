Task 1: Build a Complete NLP Preprocessing Pipeline

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return tokens
text = "Hello!!! My name is Abu. I am learning NLP..."
result = preprocess_text(text)
print(result)
```

Task 2: Detect NLP Challenges in Sentences

Detailed Explanation
1. I saw the man with the telescope.
Challenge: Ambiguity

Why?

Two meanings are possible:

I used a telescope to see the man.
The man had a telescope.
NLP models may struggle to determine the correct meaning.

2. Oh wonderful, my internet stopped during the interview.
Challenge: Sarcasm

Why?

The word:

wonderful
is positive.

But:

internet stopped during interview
is negative.

Humans understand the sarcasm, but many NLP models may misclassify it as positive sentiment.

3. He kicked the bucket yesterday.
Challenge: Context (Idiom)

Why?

Literal meaning:

He kicked a bucket.
Actual meaning:

He died.
The model must understand the phrase as an idiom, not word-by-word.

4. BRB, I need to answer a call.
Challenge: Abbreviation/Slang

Why?

BRB means:

Be Right Back
Without slang knowledge, an NLP model may not understand the sentence correctly.

5. Chat is my favorite animal.
Challenge: Ambiguity

Why?

The word "Chat" has multiple meanings:

Conversation
Chatbot
Animal name in some languages
Without context, the intended meaning is unclear.

Interview Questions & Answers
What is Ambiguity?
Answer:

A sentence having more than one possible meaning.

Example:

I saw the man with the telescope.
What is Sarcasm?
Answer:

Using positive words to express a negative feeling.

Example:

Wonderful! My phone battery died.
What is Context?
Answer:

The surrounding meaning that helps understand a sentence correctly.

Example:

He kicked the bucket.
Context tells us it means "died."

What is Slang/Abbreviation?
Answer:

Shortened informal words used in communication.

Examples:

BRB = Be Right Back



LOL = Laugh Out Loud



OMG = Oh My God

Task 3: Build BoW and TF-IDF

```python

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string


nltk.download('punkt')
nltk.download('stopwords')

# Documents
docs = [
    "I love cricket",
    "I love football",
    "Cricket is my favorite sport"
]


stop_words = set(stopwords.words('english'))

# Part A: Preprocessing


processed_docs = []

for doc in docs:
    # Lowercase
    doc = doc.lower()

    # Tokenization
    tokens = word_tokenize(doc)

    # Remove punctuation
    tokens = [word for word in tokens if word not in string.punctuation]

    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    processed_docs.append(tokens)

print("Preprocessed Documents:")
print(processed_docs)

# Part B: Vocabulary

vocabulary = sorted(set(word for doc in processed_docs for word in doc))

print("\nVocabulary:")
print(vocabulary)

# Part C: Bag of Words

processed_text = [' '.join(doc) for doc in processed_docs]

bow = CountVectorizer(vocabulary=vocabulary)
bow_matrix = bow.fit_transform(processed_text)

print("\nBag of Words Matrix:")
print(bow_matrix.toarray())

print("\nFeature Names:")
print(bow.get_feature_names_out())

# Part D: TF-IDF

tfidf = TfidfVectorizer(vocabulary=vocabulary)
tfidf_matrix = tfidf.fit_transform(processed_text)

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

# Display TF-IDF Scores

print("\nTF-IDF Scores:")

feature_names = tfidf.get_feature_names_out()

for i, doc in enumerate(tfidf_matrix.toarray()):
    print(f"\nDocument {i+1}")
    for word, score in zip(feature_names, doc):
        print(f"{word}: {round(score, 3)}")
```
